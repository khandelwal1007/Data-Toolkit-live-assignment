```python
# 14 . Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', and 'Humidity'. Perform the following tasks using Matplotlib: - Plot 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity'). - Label the x-axis as 'Date'. - Set the title of the plot as 'Temperature and Humidity Over Time'.


#Generate a time-series dataset with 30 days.
dates = pd.date_range(start = '2024-09-01', periods = 30)

#Generate random data for temperature and Humidity
np.random.seed(0) #for reprodcibility
temperature = np.random.randint(15, 35, size = 30)
humidity = np.random.randint(40, 80, size = 30)

#Create a dataframe

data = pd.DataFrame({'Date':dates,
                     'Temperature':temperature,
                     'Humidity':humidity})


#plotting the data
fig, ax1 = plt.subplots()


#plot Temperature on the left y-axis 
ax1.plot(data['Date'], data['Temperature'], color = 'tab:red', label = "Temperature")
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis = 'y', labelcolor = 'tab:red')


#create another y-axis for humidity on the rigth 
ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color = 'tab:blue', label = "Humidity")
ax2.set_ylabel('Humidity (%)', color = 'tab:blue')
ax2.tick_params(axis = 'y', labelcolor = 'tab:blue')



#set the title
plt.title("Temperature and Humidity over time")

#Display the plot
plt.show()





```

```python
#  17. - Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(0)
x=np.random.randn(100)
y=np.random.randn(100)
# create dataframe
df= pd.DataFrame({"x":x,"y":y})
# Define the quadrant for each point based on the sign of x and y
def get_quardrant(x,y):
    if x>0 and y>0:
        return "Q1"
    elif x<0 and y>0:
        return "Q2"
    elif x<0 and y<0:
        return "Q3"
    elif x>0 and y<0:
        return "Q4"
# apply
df["Quardrant"]=df.apply(lambda row: get_quardrant(row["x"],row["y"]), axis=1)


sns.relplot(data=df,x="x",y="y",hue="Quardrant")
plt.xlabel("X_axis")
plt.ylabel("Y-axis")
plt.title("Quardrant-wise scatter plot")
plt.show()


```

```python
# 18.Using Bokeh, plot a line chart of a sine wave function. Add grid lines, label the axes, and set the title as 'Sine Wave Function'.



from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np

# Generate sine wave data
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Prepare the output file (for displaying in a browser)
output_file("sine_wave.html")

# Create a new plot with title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with a legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

```python
# 19.Using Bokeh, generate a bar chart of randomly generated categorical data. Color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'


from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd
import numpy as np

categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(10, 100, size=len(categories))
df = pd.DataFrame({'Category': categories, 'Value': values})
source = ColumnDataSource(df)

output_file("categorical_bar_chart.html")

p = figure(x_range=df['Category'], title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value', height=400, width=600)

p.vbar(x='Category', top='Value', width=0.9, source=source, color='blue')

hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.outline_line_color = None
p.xaxis.major_label_orientation = 1

show(p)
```

```python
# 20.Using Plotly, create a basic line plot of a randomly generated dataset.
#Label the axes and set the title as 'Simple Line Plot



import plotly.graph_objects as go
import numpy as np

x = np.linspace(0, 10, 100)
y = np.random.random(100)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))

fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

fig.show()
```

```python
# 21.Using Plotly, create an interactive pie chart of randomly generated data.
#Add labels and percentages, and set the title as 'Interactive Pie Chart.



import plotly.graph_objects as go
import numpy as np

labels = [f'Category {i}' for i in range(5)]
values = np.random.randint(10, 100, size=len(labels))

fig = go.Figure(data=[go.Pie(labels=labels, values=values, 
                             textinfo='label+percent', 
                             insidetextorientation='radial')])

fig.update_layout(title='Interactive Pie Chart')

fig.show()
```
